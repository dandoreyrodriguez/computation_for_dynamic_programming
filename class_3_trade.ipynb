## Two endowment goods, Two countries, One period, Free Trade

The most basic reason to trade is because two countries have goods the other doesn't. Imagine there is a single period. There are two countries, $i \in \{0, 1\}$. There are also two goods, $j \in \{0,1\}$. Notice, with the zero-indexing we already have an eye to implementing this in python. Suppose the consumers are identical. We will introduce a device known as the Armington aggregator, an ingenious help. Consumers from each country will have aggregate consumption $C_{(i)}$. We don't have time subscripts because there is only one period. Aggregate consumption is defined as a function of consumption of individual goods,
$$
C_{(i)} = \left[a_{(i),0}^{\frac{1}{\sigma}}c_{(i),0}^{\frac{\sigma-1}{\sigma}} + a_{(i),1}^{\frac{1}{\sigma}}c_{(i),1}^{\frac{\sigma-1}{\sigma}} \right]^{\frac{\sigma}{\sigma-1}}
$$
where the parameters are restricted such that $\sigma>0$ and $a_{(i),0}+a_{(i),1}=1$ for each country. Each consumer will have a utility function $u(C_{(i)})$ over the aggregate good. Let $u$ have the standard properties.

Let's call $I_{(i)}$ the income of country $i$. Suppose $I_{(0)} = p_0 y_0$, where $p_0$ is the price of good 0 and $y_0$ is country 0's endowment of that good. Similarly, let $I_{(1)} = p_1 y_1$. Suppose $y_0, y_1, p_0, p_1 > 0$. So each country only has one of the two goods.

Each country's equilibrium decision is comprised of two choices: 

1. Choose aggregate consumption, $C_{(i)}$, subject to the aggregate budget constraint $P_{(i)} C_{(i)}$. $P_{(i)}$ is the price of one unit of the aggregate good.
2. Choose aggregate consumption $C_{(i)}$, choose consumption bundle $(c_{(i),0}, c_{(i),1})$ subject to the budget constraint.

We are doing this with an eye to the fact that 2 is an intratemporal choice...this will help later. Problem 1 seems trivial in a one-period set-up: each country will consume their entire endowment. $C_{(i)} = p_i y_i$. 

In the absence of tariffs, step 2 can be written as a cost minimisation (or maximise the negative of costs): 

$$
\max_{\{c_{(i),0}, c_{(i),1}\}}  - p_0 c_{(i),0} - p_1 c_{(i),1}  \quad \text{s.t.} \quad \left[a_{(i),0}^{\frac{1}{\sigma}}c_{(i),0}^{\frac{\sigma-1}{\sigma}} + a_{(i),1}^{\frac{1}{\sigma}}c_{(i),1}^{\frac{\sigma-1}{\sigma}} \right]^{\frac{\sigma}{\sigma-1}} = C_{(i)}
$$

If $\lambda_{(i)}$ is the Lagrangian from the maximisation, the first order conditions are: 

$$
p_j = \lambda_{(i)}  C_{(i)}^{\frac{1}{\sigma - 1}}  a_{(i),j}^{\frac{1}{\sigma}}  c_{(i),j}^{-\frac{1}{\sigma}}
$$

Plug this into the expression for expenditure, $p_0 c_{(i),0} + p_1 c_{(i),1} = I_{(i)}$:

$$
I_{(i)} 
= \lambda_{(i)} C_{(i)}^{\frac{1}{\sigma - 1}} \left( a_{(i),0}^{\frac{1}{\sigma}} c_{(i),0}^{\frac{\sigma - 1}{\sigma}} + a_{(i),1}^{\frac{1}{\sigma}} c_{(i),1}^{\frac{\sigma - 1}{\sigma}}  \right)
$$

The final term is just the inner expression of the CES aggregator. Hence:

$$
\lambda_{(i)} C_{(i)}^{\frac{1}{\sigma - 1}}  C_{(i)}^{\frac{\sigma - 1}{\sigma}} = \lambda_{(i)} C_{(i)} = I_{(i)}
$$

Thus, the Lagrange multiplier $\lambda_{(i)}$ is the income required to purchase one unit of the aggregate good. It is an aggregate price index, $P_{(i)} = \lambda_{(i)}$. To get a clean expression for $P_{(i)}$, set $p_0 c_{(i),0} + p_1  c_{(i),1} = P C$ and set $C=1$ everywhere it appears. That gives,

$$
P_{(i)} = \left( p_0^{1-\sigma}a_{(i),0} + p_1^{1-\sigma}a_{(i),1} \right)^{\frac{1}{1-\sigma}}
$$

Plugging this into the FOC one gets:

The easiest way to get Hicksian demand is to invoke Shepard's lemma. The expenditure function (where $C_{(i)}$ takes the place of a target utility) is just $P_{(i)} C_{(i)}$. So, 

$$
h_{(i),j}(p_0,p_1,C_{(i)})
= C_{(i)} \frac{\partial P_{(i)}}{\partial p_j}
= C_{(i)} \underbrace{a_{(i),j} \left(\frac{p_j}{P_{(i)}}\right)^{-\sigma}}_{\text{Share of agg. consumption}}
$$

This is a classic Walrasian equilibrium. Equilibrium here is a set of prices $\{1, p_1\}$ (remember we need a numeraire, let it be good 0) and consumption plans $\{C_{(i)}, c_{(i),0}, c_{(i),1}\}_{i=0}^1$ such that:
- Consumers solve their constrained optimisation
- Markets clear, $y_j = c_{(0),j}+c_{(1),j}$

In this two-good, two-country example this boils down to checking markets clear in good 1. Walras' law will clear 0. Realise that when good 0 is fixed, $P_{(i)}(p_1)$ is just a function of the "outstanding" good. The market clearing condition after substituting relevant expressions is:

$$
y_1 = y_0 a_{(0),1} p_1^{-\sigma} \left(\frac{1}{P_{(0)}(p_1)}\right)^{1-\sigma}+y_1 a_{(1),1} p_1^{1-\sigma}\left(\frac{1}{P_{(1)}(p_1)}\right)^{1-\sigma}
$$

Plugging in the index this is,

$$
y_1 = \frac{y_0 a_{(0),1} p_1^{-\sigma}}{a_{(0),0}+a_{(0),1}p_1^{1-\sigma}}+\frac{y_1 a_{(1),1} p_1^{1-\sigma}}{a_{(1),0}+a_{(1),1}p_1^{1-\sigma}}
$$

Notice that this is, in general, a non-linear equation in $p_1$. This is why computational methods are so important in economics. Theory is beautiful in helping us get to equilibrium conditions. However, the set of equilibria which can be solved by pen and paper is small. I will write a little code which will help me solve this example.


```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.optimize import brentq

def solve_two_country_trade(y_0, y_1, a_00, a_01, a_10, a_11, sigma, ax=None):
    
    def P_0_index(p_1):
        P_0_inside = a_00 + a_01 * p_1**(1-sigma)
        P_0 = P_0_inside**(1/(1-sigma))
        return P_0
    
    def P_1_index(p_1):
        P_1_inside = a_10 + a_11 * p_1**(1-sigma)
        P_1 = P_1_inside**(1/(1-sigma))
        return P_1
   
    def h_01_trade(p_1):
        income = y_0
        P_0 = P_0_index(p_1)
        c_01 = income * a_01 * p_1**(-sigma) * (1/P_0)**(1-sigma)
        return c_01

    def h_11_trade(p_1):
        income = p_1 * y_1
        P_1 = P_1_index(p_1)
        c_11 = income * a_11 * p_1**(-sigma) * (1/P_1)**(1-sigma)
        return c_11

    def good1_excess_demand(p_1):
        c_01 = h_01_trade(p_1)
        c_11 = h_11_trade(p_1)
        return c_01 + c_11 - y_1

    # Solve for equilibrium price
    p_1 = brentq(good1_excess_demand, 0.1, 10)

    # Price indices
    P_0 = P_0_index(p_1)
    P_1 = P_1_index(p_1)

    # Aggregate consumption
    C_0 = y_0 / P_0
    C_1 = p_1 * y_1 / P_1

    # Allocations
    c_01 = h_01_trade(p_1)
    c_00 = P_0 * C_0 - p_1 * c_01
    c_11 = h_11_trade(p_1)
    c_10 = P_1 * C_1 - p_1 * c_11

    if ax is not None:
        ax.plot([y_0], [0], 'ro', label='Country 0 endowment', markersize=10)
        ax.plot([0], [y_1], 'bo', label='Country 1 endowment', markersize=10)
        ax.plot([c_00], [c_01], 'rx', label='Country 0 final bundle', markersize=10, markeredgewidth=2)
        ax.plot([c_10], [c_11], 'bx', label='Country 1 final bundle', markersize=10, markeredgewidth=2)
        ax.plot([0, y_0], [y_1, 0], 'k--', label='Endowment line')
        ax.set_xlim(0, y_0 * 1.2)
        ax.set_ylim(0, y_1 * 1.2)
        ax.set_xlabel('Good 0')
        ax.set_ylabel('Good 1')
        ax.legend()
        ax.set_title(rf'$y_0=$ {y_0}, $y_1=${y_1}, $p_1=${p_1:.2f}')
        ax.grid(True)

# ---- Run two examples and plot side by side ----
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Symmetric endowment
solve_two_country_trade(
    y_0=2, y_1=2,
    a_00=0.5, a_01=0.5,
    a_10=0.5, a_11=0.5,
    sigma=2,
    ax=ax1
)
 
# Asymmetric endowment
solve_two_country_trade(
    y_0=2, y_1=4,
    a_00=0.5, a_01=0.5,
    a_10=0.5, a_11=0.5,
    sigma=2,
    ax=ax2
)

plt.tight_layout()
plt.show()
```

Notice that even this completely static problem wasn't trivially easy. Nevertheless, we introduced devices which are going to help us. You can't understand the fancy stuff if you don't understand the basics. Moreover, the way we set the problems up will make jumping to the $N$ country case easier. But I'm getting ahead of myself...

## Two endowment goods, Two countries, One-period, Tariffs

Now we begin our study of tariffs. We do so with an eye to the future, we want a set-up which will scale up easily (CRS?) as we add countries, varieties, and dynamics. The only change we make to the set-up in the previous model is to introduce ad valorem tariffs. They drive a wedge between the world price (price at the border) and the price consumers pay.

In the current set up, country 0 only has good 0 and country 1 only has good 1. Therefore, there are two non-zero tariffs. Let $\tau_{(0),1}$ be the ad-valorem tariff country 0 puts on good from the only outsider, country 1. Likewise $\tau_{(1),0}$ is the corresponding tariff. 

Let's break the problem into two steps as before. We will do this for country 0, but the same holds in reverse for country 1. 

First, it is clear that aggregate consumption is $C_{(0)} = I_{(0)}$. This time, income is, $I_{(0)} = p_0 y_0 + T_{(0)}$. That is, there is a national tariff revenue which is to be rebated but which is lump sum from the point of view of an agent. In equilibrium, this will be $T_0 = \tau_{(0),1} c_{(0),1}$. This is because the aggregate national tariff revenue is assumed to be rebated. It is very important to remember that $c_{(0),1}$ is aggregate national consumption. An individual chooses only their consumption, and takes national consumption as given. That is, an individual does not have a personal effect on aggregate tariff revenues. This point often used to confuse me, because in the next step we will use the same notation. The representative agent in a unit mass will consume the same amount, but this distinction is often forgotten. Don't be like me! With this in mind, the second problem for the representative *individual* in country 0 is, 


$$
\max_{\{c_{(0),0}, c_{(0),1}\}}  - p_0 c_{(0),0} - p_1(1+\tau_{(0),1}) c_{(0),1}  \quad \text{s.t.} \quad \left[a_{(0),0}^{\frac{1}{\sigma}}c_{(0),0}^{\frac{\sigma-1}{\sigma}} + a_{(0),1}^{\frac{1}{\sigma}}c_{(0),1}^{\frac{\sigma-1}{\sigma}} \right]^{\frac{\sigma}{\sigma-1}} = C_{(0)}
$$

The crucial insight here is that if you let $p_1(1+\tau_{(0),1}) = \tilde p_1$, you basically have an identical problem! This is the deep sense in which tariffs are a distortion to prices. So, the price index will be, 

$$
P_{(0)} = \left( p_0^{1-\sigma}a_{(0),0} + \tilde{p}_1^{1-\sigma}a_{(0),1} \right)^{\frac{1}{1-\sigma}}
$$

Hicksian demand is, 

$$
h_{(0),j}(p_0,\tilde{p}_1, C_{(0)})
= C_{(i)} \frac{\partial P_{(i)}}{\partial p_j}
= C_{(0)} \underbrace{a_{(0),j} \left(\frac{\tilde{p}_j}{P_{(0)}}\right)^{-\sigma}}_{\text{Share of agg. consumption}}
$$


This is a tariff-specific Walrasian equilibrium. For a set of tariffs and endowments, equilibrium here is a set of prices $\{1, p_1\}$ and consumption plans $\{C_{(i)}, c_{(i),0}, c_{(i),1}\}_{i=0}^1$ such that:
- Consumers solve their constrained optimisation
- Markets clear, $y_j = c_{(0),j}+c_{(1),j}$
- Tariff revenue is fully rebated within period

In this two-good, two-country example this boils down to checking markets clear in good 1. Walras' law will clear 0. Realise that when good 0 is the numeraire, $P_{(i)}(p_1)$ is just a function of the good 1 equilibrium price, tariffs and preferences are known. 

$$
y_1
= h_{(0),1}(\tilde{p}_1, C_{(0)})+h_{(1),1}(C_{(0)})
$$

Notice that this is going to be a little bit trickier than before. This is because of aggregate consumption, which depends on income which in turn depends on the quantity of the tariff good demanded---since that impacts the size of the tariff rebate which boosts income. Before, income depended only on the endowment (and $p_1$ for country 1). The easiest way forward here is to realise that in equilibrium, 

$$
c_{(0),1} = \frac{y_0 + p_1 \tau_{(0),1} \, c_{(0),1}}{P_{(0)}(p_1)} \cdot a_{(0),1} \left( \frac{(1+\tau_{(0),1}) p_1}{P_{(0)}(p_1)} \right)^{-\sigma}
$$

Likewise for country 1, 

$$
c_{(1),1} = \frac{p_1 y_1 + \tau_{(1),0} \, c_{(1),0}}{P_{(1)}(p_1)} \cdot a_{(1),1} \left( \frac{p_1}{P_{(1)}(p_1)} \right)^{-\sigma}
$$

That is, you can write the market clearing condition as a function of the equilibrium price, $p_1$ and all the other known parameters, i.e. endowments, tariffs, and preferences. As before, we can use numerical methods to solve for the equlibrium price. Once we have this, everything else can be computed. 

```python
# solve the tariff case
def solve_two_country_tariff(y_0, y_1, a_00, a_01, a_10, a_11, t_01, t_10,  sigma, ax=None):
    
    def P_0_index(p_1):
        P_0_inside = a_00 + a_01 * ( (1+t_01)*p_1 )**(1 - sigma)
        P_0 = P_0_inside**( 1 / (1 - sigma))
        return P_0 
    
    def P_1_index(p_1):
        P_1_inside = a_10*(1+t_10)**(1-sigma) + a_11 * p_1**(1 - sigma)
        P_1 = P_1_inside**( 1 / (1 - sigma))
        return P_1
    
    # write a function for good 1, country 0 hicksian demand
    def h_01_tariff(p_1):
        P_0 = P_0_index(p_1)
        A = a_01 * ( (1+t_01)*p_1 )**(-sigma) * (1 / P_0)**(1-sigma) 
        c_01 = y_0 * (A/(1-p_1*t_01*A))
        return c_01
    
    # write a function for good 1, country 1 hicksian demand
    def h_11_tariff(p_1):
        P_1 = P_1_index(p_1)
        B = a_11 * p_1**(-sigma) * (1/P_1)**(1-sigma)
        c_11 = (p_1 * y_1 * B) / (1 - p_1 *t_10 * B )
        return c_11
    
    # write a function for good 1, country 0 market clearing
    def good1_excess_demand(p_1):
        c_01 = h_01_tariff(p_1)
        c_11 = h_11_tariff(p_1)
        return c_01 + c_11 - y_1
    
    # Solve for equilibrium price
    p_1 = brentq(good1_excess_demand, 0.1, 10)

    # solve for allocations of good 1
    c_01 = h_01_tariff(p_1)
    c_11 = h_11_tariff(p_1)
    # price index
    P_0 = P_0_index(p_1)
    # income good 0
    I_0 = y_0 + p_1 * t_01 * c_01
    c_00 = I_0 * a_00 * (1/P_0)**(1-sigma)
    c_10 = y_0 - c_00

    if ax is not None:
        ax.plot([y_0], [0], 'ro', label='Country 0 endowment', markersize=10)
        ax.plot([0], [y_1], 'bo', label='Country 1 endowment', markersize=10)
        ax.plot([c_00], [c_01], 'rx', label='Country 0 tariff bundle', markersize=10, markeredgewidth=2)
        ax.plot([c_10], [c_11], 'bx', label='Country 1 tariff bundle', markersize=10, markeredgewidth=2)
        ax.plot([0, y_0], [y_1, 0], 'k--', label='Endowment line')
        ax.set_xlim(0, y_0 * 1.2)
        ax.set_ylim(0, y_1 * 1.2)
        ax.set_xlabel('Good 0')
        ax.set_ylabel('Good 1')
        ax.legend()
        ax.set_title(rf'$y_0=$ {y_0}, $y_1=${y_1}, $p_1=${p_1:.2f}, $\tau_{{(0),1}} = ${t_01},$\tau_{{(1),0}} = ${t_10}')
        ax.grid(True)

# ---- Run two examples and plot side by side ----
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# test run with zero tariffs and symmetric endowments
solve_two_country_tariff(
    y_0=2, y_1=2,
    a_00=0.5, a_01=0.5,
    a_10=0.5, a_11=0.5,
    t_01=0.3, t_10=0,
    sigma=2, ax=ax1
)

# test run with zero tariffs and symmetric endowments
solve_two_country_tariff(
    y_0=2, y_1=4,
    a_00=0.5, a_01=0.5,
    a_10=0.5, a_11=0.5,
    t_01=0.3, t_10=0,
    sigma=2, ax=ax2
)
```

See how this compares to the first example! Well, the main intuition we gained whilst doing the derivations: tariffs distort relative prices, altering the cost of substituting from one good to another. Income is replenished, so there is no income effect of tariffs, only a substitution effect. Remember, there is no production in this set-up.

Our next step is to generalise to $N$ countries. This will need to involve linear algebra. Doing multi-country trade is to use the same apparatus as usual, but with linear algebra. It may appear tedious at first. But as we generalise, the linear algebra will give us more and more mileage. 

## N endowment goods, N countries, One Period, Tariffs

The set-up will be identical as the two-country example, though this time we will generalise and scale the results using linear algebra. This is where python really shines. Suppose that $N$ countries are endowed with $N$ distinct goods. Country $i\in \{0,N-1\}$ has the Armington Aggregator, 

$$
C_{(i)} = \bigg[ \sum_{j=0}^{N-1} a_{i,j} c_{i,j}^{\frac{\sigma-1}{\sigma}}\bigg]^{\frac{\sigma}{\sigma-1}}
$$

But remember, we want to use linear algebra to make our lives easier. Let $\mathbf{c}$ be a $N\times1$ vector of aggregate consumption across the $i$ countries, where the $i$ th country's consumption is the entry in the $i$ th in the vector. Let $\mathbf{C}$ be a $N \times N$ matrix, where the $(i,j)$ th entry refers to country $i$ 's consumption of country $j$ 's good. Let $\mathbf{A}$ be an $N \times N$ matrix which captures the preference weights. I will define the power of a matrix $\mathbf{x}^a$ to be element-wise. That is, $[\mathbf{X}^a]_{ij} := x_{ij}^a$. Let $\mathbf{i}$ be an $N$ dimensional vector of ones, the Armington aggregator can be summarised as, 

$$
\mathbf{c}
= \bigg[ \bigg( \mathbf{A} \circ \mathbf{C}^{\frac{\sigma-1}{\sigma}} \bigg) \, \mathbf{i}\bigg]^{\frac{\sigma}{\sigma-1}}
$$

As before, the problem has two steps. First, the consumer chooses aggregate consumption. Since there is one period they consume their income. Suppose that $\mathbf{Y}$ is a $N \times N$ matrix of endowments, where the $i$ th row refers to that country's endowment of the various goods. Aggregate income is, 

$$ 
\mathbf{I}
= \mathbf{Y} \mathbf{p} + \mathbf{t}
$$

$\mathbf{p}$ is a $N$ dimensional vector of prices, and $\mathbf{t}$ is the $N$ dimensional vector of tariff rebates. 

As before, the second step will be to choose the bundle of goods subject to aggregate consumption. We need the set of prices faced by a consumer after tariffs. Let $\mathbf{\tau}$ be a $N \times N$ matrix, where the $(i,j)$ th entry refers to the tariff country $i$ places on country $j$. Then, 

$$
\tilde{\mathbf{P}} 
= \mathbf{i} \,  \mathbf{p}' \circ (1+\mathbf{\tau})
$$

$\tilde{\mathbf{P}}$ is a $N \times N$ matrix of consumer prices describing the post-tariff prices each country's consumers face. The vector of the agg price indices can be given as, 

$$
\mathbf{p}_{\text{agg}}
= \bigg[ \bigg( \mathbf{A} \circ \tilde{\mathbf{P}}^{1-\sigma} \bigg) \, \mathbf{i} \bigg]^{\frac{1}{1-\sigma}}
$$

We will define element-wise division with the $\div$ operator. Let $\mathbf{R} := \tilde{\mathbf{P}} \div ( \mathbf{p}_{\text{agg}} \, \mathbf{i}' )$ be a $N \times N$ matrix capturing the relative price of the good versus the consumer basket. 

As before, the Hicksian demand function is, 

$$
h(\tilde{\mathbf{P}}, \mathbf{c})
= \mathbf{c} \, \mathbf{i}' \circ \mathbf{A} \circ \mathbf{R}^{-\sigma}
$$

For a given set of endowments, tarffs, and preferences, an equilibrium is a set of consumption choices $\{\mathbf{c}, \mathbf{C}\}$ and prices $\mathbf{p}$ such that:
- Consumer solve their two problems (hicksian demand and budget constraints satisfied)
- Markets clear, $\mathbf{y} = \mathbf{C} \mathbf{i} $
- The government rebates all tariff revenue, $\mathbf{t} = ( \mathbf{C} \circ \mathbf{\tau} ) \,  \mathbf{p} $

Remember that, $\mathbf{I} = \mathbf{p}_{\text{agg}} \circ \mathbf{c}$. Just as last time, the best way to solve is to write the market clearing condition as a function in $\mathbf{p}$. As last time, 

$$
\mathbf{C} 
= \mathbf{c} \, \mathbf{i}' \circ \mathbf{A} \circ \mathbf{R}^{-\sigma} 
= \bigg[ (\mathbf{Y} \mathbf{p} + \underbrace{( \mathbf{C} \circ \mathbf{\tau} ) \,  \mathbf{p}}_{\mathbf{t}} ) \div \mathbf{p}_{\text{agg}} \bigg] \, \mathbf{i}' \circ \mathbf{A} \circ \mathbf{R}^{-\sigma}
$$

This is trickier than the previous case. A country's consumption of the $j$ th good depends on the tariff income which depends on its consumption of all other goods. So we cannot easily isolate $\mathbf{C}$. But here we can apply the same logic as in dynamic programming, even if it is a static problem. Here will be our algorithm for solving the equilibrium. 

1. Guess $\mathbf{p}$
2. Solve for a fixed point $\mathbf{C}$
3. Check if excess demand
4. If excess (insufficient) demand in a good, increase (decrease) the price and return to 1. If no excess demand (within tolerance), stop. 

```python
from  numba import njit
from scipy.optimize import root

# compute the agg price index 
@njit
def price_index(p, tau, A, sigma):
    p = p.reshape(-1,1)
    N = len(p)
    P_tilde = (np.ones((N,1)) @ p.T) * (1 + tau)
    p_agg_inside = (A * (P_tilde**(1-sigma))) @ np.ones((N,1))
    p_agg = p_agg_inside ** (1/(1-sigma))
    return p_agg

@njit
def agg_cons(p, C, tau, Y, A, sigma):
    p = p.reshape(-1,1)
    tar_rev = ( (C * tau) @ p ) 
    agg_cons = ( (Y @ p) + tar_rev ) / price_index(p, tau, A, sigma)
    return agg_cons

# compute hicksian demand 
@njit
def hicks(c_agg, A, p, tau, sigma):
    c_agg = c_agg.reshape(-1,1)
    p = p.reshape(-1,1)
    N = len(p)
    P_tilde = (np.ones((N,1)) @ p.T) * (1 + tau)
    R = P_tilde / (price_index(p, tau, A, sigma) @ np.ones((1,N)))
    C = ( c_agg @ np.ones((1,N))) * A * (R**(-sigma)) 
    return C

# write function which solves for C 
@njit
def solve_C(p, A, tau, Y, sigma, tolerance = 1e-16):
    p = p.reshape(-1,1)
    N = len(p)
    # initial guess
    C_old = np.ones((N,N))
    # initialise the error
    error = 1 + tolerance
    # start the loop 
    while error > tolerance:
        # calculate agg consumption
        c_agg = agg_cons(p, C_old, tau, Y, A, sigma)
        # update C using hicksian demand function
        C_new = hicks(c_agg, A, p, tau, sigma)
        # check error
        error = np.max(np.abs(C_new - C_old))
        # update 
        C_old = C_new.copy()
    # return consumption object
    return C_new

# write a function which computes excess demand
@njit
def excess_demand(p, A, tau, Y, sigma):
    p = p.reshape(-1,1)
    N = len(p)
    C = solve_C(p, A, tau, Y, sigma)
    demand = C.T @ np.ones((N,1))
    supply = Y.T @ np.ones((N,1))
    return demand - supply 

# make a class for these types of cases
class tariff_one_period:

    def __init__(self, A, tau, Y, sigma):
        # set up pars
        self.A = A
        self.tau = tau
        self.Y = Y
        self.sigma = sigma 
        self.N = A.shape[0]
        # catch results
        self.p_eq = None
        self.p_agg = None
        self.c_agg = None
        self.C = None

    def solve_for_eq(self):
        # set up friendly object for solving
        def obj(p_log_rest):
            p_log_full = np.concatenate(([0.0], p_log_rest))
            p_full = np.exp(p_log_full).reshape(-1,1)
            # get excess demand
            z = excess_demand(p_full, self.A, self.tau, self.Y, self.sigma)
            # drop the first good as we want to fix that price (the numeraire)
            z = z[1:]
            # return a flattened z, which is what scipy will expect
            return z.flatten()
        # guess the first 
        p_log_rest_guess = np.zeros(self.N-1)
        sol = root(obj, p_log_rest_guess, method = 'hybr')
        if not sol.success:
            raise RuntimeError("Equilibrium price solver failed:", sol.message)
        # reconstruct full prices
        p_log_full_eq = np.concatenate(([0.0], sol.x))
        p_eq = np.exp(p_log_full_eq).reshape(-1,1)
        # catch results 
        self.p_eq = p_eq
        self.C = solve_C(p_eq, self.A, self.tau, self.Y, self.sigma)
        self.c_agg = agg_cons(p_eq, self.C, self.tau, self.Y, self.A, self.sigma)
        self.p_agg = price_index(p_eq, self.tau, self.A, self.sigma)


Y = 2 * np.eye(10)
A = 0.5 * np.ones((10,10))
sigma = 2
#Initialize tariff matrix
tau = np.zeros((10, 10))
# Country 0 taxes imports from 1 and 2
tau[0, 1] = 0.3   # 30% tariff on good 1
tau[0, 2] = 0.2   # 20% tariff on good 2
# Country 1 protects against 0 and 3
tau[1, 0] = 0.25
tau[1, 3] = 0.4
# Country 2 retaliates against country 1
tau[2, 1] = 0.2


multi_tariff_0 = tariff_one_period(A, tau, Y, sigma)
multi_tariff_0.solve_for_eq()

import matplotlib.pyplot as plt

plt.imshow(multi_tariff_0.C, cmap="viridis")
plt.colorbar(label="Quantity")
plt.xlabel("Good origin (j)")
plt.ylabel("Consumer country (i)")
plt.title("Consumption matrix C")


```

## S sectors, N countries, Production goods, One-period, Tariffs

So far we have dealt only with endowment goods. This has been useful, but tariffs impact producers as much as they impact consumers. The reason we remained with endowment goods for so long is that production introduces a quirk which is subtle. In the endowment good case, multiple countries can be endowed with the same good. Suppose the good is cars. It seems sensible that multiple countries should like to sell cars, at home and abroad. However, as soon as we move intro production, the most productive producer will come to dominate that sector. More specifically, we will want to use constant returns to scale (CRTS) technology. In this case, the most productive producer will simply scale up to become the only producer of the good. To keep CRTS but allow multiple countries to produce cars in equilibrium, we will need to introduce some variation between American and Japanese cars. This is OK because we have learned how to deal with variation beneath an aggregate good: use the Armington aggregator!

Secondly, we want richness in production. The production networks literature provides a way to do just that. This literature supposes that firms use inputs from other sectors. In closed-economy set-ups, each good is interpreted as being a sector. In our trade set-up, we will need country-sector pairs producing a unique good. 

Suppose there are $N$ countries. Suppose also there are $S$ sectors. Consumers have two Armington aggregators, one over sectors, and one over varieties. That is, a consumer from country $i$ experiences utility $u(C_i)$ from consuming $C_i$ of the aggregate good. This aggregate goods is comprised of sectors as follows: 

$$
C_i = \bigg[ \sum_{s = 0}^{S-1} a_{is}^{\frac{1}{\sigma_i}} C_{is}^{\frac{\sigma_i - 1}{\sigma_i}}\bigg]^{\frac{\sigma_i}{\sigma_i-1}}
$$

Each sectoral bundle $C_{is}$ is itself an Armington aggregator over unique country-sector varieties within sector $s$:

$$
C_{is} = \left[ \sum_{n = 0}^{N-1} b_{isn}^{\frac{1}{\rho_s}} \cdot c_{isn}^{\frac{\rho_{is} - 1}{\rho_{is}}} \right]^{\frac{\rho_{is}}{\rho_{is} - 1}}
$$

Here:
- $c_{isn}$ is country $i$ 's consumption of sector $s$'s variety from country $n$
- $a_{is}, b_{isn}$ are preference parameters
- $\sigma_i$ is country $i$ 's elasticity across sectors. We will store it in a $N$-dimensional vector $\mathbf{\sigma}$
- $\rho_{is} > 1$ is the elasticity of substitution between varieties within sector $s$. We store it in a $N \times S$ matrix $\mathbf{\rho}$

As last time, we will want to break the problem down into steps. 
1. The consumer chooses aggregate consumption, $C_i$ subject to the budget constraint
2. The consumer chooses their sectoral split, $C_{is}$, by minimising expenditure subject to achieving the aggregate consumption target
3. The consumer chooses their variety split, $c_{isn}$, by minimising expenditure subject to achieving the sectoral consumption target

The problem is significantly harder as a computational matter. Now there are $N \times S$ consumption choices to make, with as many prices. There will be an aggregate price level (something like CPI), there will be a sectoral aggregate price level, and prices for country-variety pairs. However the expressions are exactly the same as before! So we don't need to rederive anything, we just need to keep track of labels carefully. 

1. Aggregate consumption

As before, $P_i C_i = I_i $. We will define income once we have looked at the production side (and dealt with wage income). So, across countries, this can again be written simply as, 

$$
\mathbf{p} \circ \mathbf{c} = \mathbf{I}
$$

I wish it were always that easy. This is an $N \times 1$ vector. Each row is a country. 

2. Sectoral consumption

This will be very similar to before. Instead of goods, we have sectors. Tariffs do not feature directly here but $\mathbf{P}_{\text{sec}}$ is an $N \times S$ matrix which captures each country's sectoral price index, which in turn captures tariffs. That is, the $i$ th row captures the price index in that country for each sector. So, sectoral hicksian demand can be written, 

$$
h_{\text{sec}}(\mathbf{P}_{\text{sec}}, \mathbf{c})
= \mathbf{c} \, \mathbf{i}' \circ \mathbf{A} \circ \mathbf{R}_{\text{sec}}^{-\mathbf{\sigma}}
$$

$\mathbf{R}_{\text{sec}} := \mathbf{P}_{\text{sec}} \div \mathbf{p}_{\text{agg}} \mathbf{i}'$, where division is element-wise as before. Importantly, however, we need to define what it means to raise a matrix to a power of a vector. I define is at row-wise. In other words, $[\mathbf{R}_{\text{sec}}^{\mathbf{\sigma}}]_{is} := R_{is}^{\sigma_i} $. 

This expression is familiar, and gives a $N \times S$ matrix of sectoral Hisckian demands. Analgously to before, the aggregate price index is a function of the sectoral prices indices. Specifically, 

$$
\mathbf{p}
= \bigg[ \bigg( \mathbf{A} \circ {\mathbf{P}_{\text{sec}}}^{1-\sigma} \bigg) \, \mathbf{i} \bigg]^{\frac{1}{1-\sigma}}

$$

3. Variety consumption

This is where things will get a little trickier. The point of using linear algrebra is to keep track of variables in one place. This was intuitive in step 2, where $h_{\text{sec}}(\mathbf{P}_{\text{sec}}, \mathbf{c})$ generally treats rows as country's and columns as sectors. As we move to consider varities, this will no longer be possible. We will aim for a $NS \times N$ dimensional hicksian demand matrix. The rows will represent a given country's consumption of a certain sector, and the columns will represent the varieties produced by each country. This will require stacking some of the quantities computed. 

We will use the following stacking transformations:

Country consumption per sector will be $\tilde{\mathbf{C}}_{\text{sec}} := \text{vec} (\mathbf{C}_{\text{sec}}')$, a $NS \times 1$ vector, where the first loop is over sectors, the second over countries. The $N+1$ th entry is the first of second country.

Similarly, we stack the sector-preferences weights to be $\tilde{\mathbf{A}}_{\text{sec}} := \text{vec} (\mathbf{A}_{\text{sec}}')$ a $NS \times 1$ vector, as we do with $\mathbf{\rho}$ to get $\tilde{\mathbf{\rho}}$. 

Finally, notice we never define how we would store the $b_{isn}$ weights. Let $\tilde{\mathbf{B}}$ house these weights in a $NS \times S$ matrix. Rows are country sector pairs, with country's appearing in fixed batches.

Finally, we need to define our price objects. Following the convention, $\tilde{\mathbf{\tau}}$ is a $NS \times N$ matrix with each country's tariff rate on a country-sector pair. Let $\tilde{\mathbf{p}}_{\text{var}}$ be a $NS \times N$ matrix holding the world prices. Finally, we need to stack the sectoral aggregate prices. So, $\tilde{\mathbf{P}}_{\text{sec}} := \text{vec} (\mathbf{P}_{\text{sec}}')$ is also $NS \times 1$. As usual, let's define a helper relative price variable, $\tilde{\mathbf{R}}_{\text{var}} := (\tilde{\mathbf{P}}_{\text{var}} \circ (1 + \tilde{\boldsymbol{\tau}})) \div  \tilde{\mathbf{P}}_{\text{sec}} \mathbf{1}_N'$. 

This hicksian demand function is, 

$$
\underbrace{h_{\text{var}}(\mathbf{P}_{\text{var}}, \mathbf{c})}_{NS \times N}
= \tilde{\mathbf{C}}_{\text{sec}} \, \mathbf{i}' \, \circ \tilde{\mathbf{B}} \circ \tilde{\mathbf{R}}_{\text{var}}^{\tilde{\mathbf{\rho}}} 
$$
